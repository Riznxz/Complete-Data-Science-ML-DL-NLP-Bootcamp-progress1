```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
X = np.array([60, 80, 100, 120, 140, 160, 180, 200, 220, 240]).reshape(-1, 1)
y = np.array([15.5, 12.8, 10.5, 8.9, 7.8, 7.2, 7.0, 6.9, 6.8, 6.7])
```

```python
poly=PolynomialFeatures(degree=3,include_bias=True)
X_poly=poly.fit_transform(X)
```

```python
model=LinearRegression()
model.fit(X_poly,y)
```

```python
X_test=np.linspace(60,240,300).reshape(-1,1)
X_test_poly=poly.transform(X_test)
y_pred=model.predict(X_test_poly)
```

```python
plt.scatter(X,y,color='r',label='actual')
plt.plot(X_test,y_pred,color='g',label='degree fit')
```
